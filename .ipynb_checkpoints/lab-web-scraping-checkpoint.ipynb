```python
rating_list = {
    "One":1,
    "Two":2,
    "Three":3,
    "Four":4,
    "Five":5
}

def get_soup(book_url: str) -> BeautifulSoup:
    resp = requests.get(book_url)
    resp.raise_for_status()
    return BeautifulSoup(resp.text, "html.parser")
```

```python
def parse_book_detail(book_url: str) -> dict:
    soup = get_soup(book_url)

    # Title
    title = soup.find("div", class_ = "product_main").h1.get_text(strip = True)

    # Price
    price_text = soup.find("p", class_ = "price_color").get_text(strip = True)
    price_text = price_text.replace("Â", "").replace("£", "")
    price = float(price_text)

    # Availability
    availability = soup.find("p", class_="instock availability").get_text(strip = True)

    # Rating(class such as 'star-rating-four')
    rating_p = soup.find("p", class_="star-rating")
    rating_class = next(c for c in rating_p["class"] if c != "star-rating")
    rating = rating_list.get(rating_class, None)

    # Genre
    breadcrumb_links = soup.select("ul.breadcrumb li a")
    genre = breadcrumb_links[-1].get_text(strip=True) if breadcrumb_links else None

    # Product info table for UPC
    upc = None
    product_table = soup.find("table", class_="table table-striped")
    if product_table:
        for row in product_table.find_all("tr"):
            heading = row.th.get_text(strip=True)
            val = row.td.get_text(strip=True)
            if heading == "UPC":
                upc = val
                break

    # Description
    description = None
    desc_header = soup.find("div", id="product_description")
    if desc_header:
        desc_p = desc_header.find_next_sibling("p")
        if desc_p:
            description = desc_p.get_text(strip=True)

    return {
        "UPC": upc,
        "Title": title,
        "Price (£)": price,
        "Rating": rating,
        "Genre": genre,
        "Availability": availability,
        "Description": description,
    }

def scrape_books(min_rating: float = 0.0, max_price: float = float("inf")) -> pd.DataFrame:
    books_data = []
    page = 1

    while True:
        page_url = book_url + f"catalogue/page-{page}.html"
        resp = requests.get(page_url)

        # Stop when page does not exist
        if resp.status_code != 200:
            break

        soup = BeautifulSoup(resp.text, "html.parser")
        book_items = soup.select("article.product_pod h3 a")

        if not book_items:
            break

        for a in book_items:
            
            relative_url = a["href"]
            
            if "catalogue/" not in relative_url:
                relative_url = relative_url.lstrip("../")
                relative_url = "catalogue/" + relative_url
            
            new_book_url = book_url + relative_url

            details = parse_book_detail(new_book_url)

            if details["Rating"] is None:
                continue
            if details["Price (£)"] is None:
                continue
            if details["Rating"] < min_rating:
                continue
            if details["Price (£)"] > max_price:
                continue

            books_data.append(details)

        page += 1

    df = pd.DataFrame(
        books_data,
        columns=[
            "UPC",
            "Title",
            "Price (£)",
            "Rating",
            "Genre",
            "Availability",
            "Description",
        ],
    )
    return df

if __name__ == "__main__":
    df_result = scrape_books(min_rating=4.0, max_price=20.0)
    print(df_result.head())
    print(f"\nTotal books scraped: {len(df_result)}")
```
